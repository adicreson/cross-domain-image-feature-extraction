```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npz file
data = np.load('evaluations.npz')

timesteps = data['timesteps']
results = data['results']

# Calculate the mean of the results across the 10 sets
mean_results = np.mean(results, axis=1)

# Close the file
data.close()

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(timesteps, mean_results, label='Mean Results')
plt.xlabel('Timesteps')
plt.ylabel('Mean Results')
plt.title('Timesteps vs. Mean Results')
plt.legend()
plt.grid(True)

# Set the x-axis limits explicitly
plt.xlim(0, 10_000_000)

plt.show()
```
